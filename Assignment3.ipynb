Try to find every color and turn them into bar-colors we could compare.

```python
number_of_pixels_of_a_color_1 = []
color_labels_1 = []
color_values_1 = []
number_of_pixels_of_a_color_2 = []
color_labels_2 = []
color_values_2 = []
number_of_pixels_of_a_color_3 = []
color_labels_3 = []
color_values_3 = []
```

```python
for icolor, rgb in enumerate(unique_1_colors):
    reds_mask = im1_data[:,:,0] == rgb[0]
    green_mask = im1_data[:,:,1] == rgb[1]
    blue_mask = im1_data[:,:,2] == rgb[2]
    
    pixel_mask = reds_mask & green_mask & blue_mask
    
    npixels = len(im1_data[pixel_mask])
    number_of_pixels_of_a_color_1.append(npixels)
    color_labels_1.append("Color #" + str(icolor))
    color_values_1.append(rgb/255)
```

```python
for icolor, rgb in enumerate(unique_2_colors):
    reds_mask = im2_data[:,:,0] == rgb[0]
    green_mask = im2_data[:,:,1] == rgb[1]
    blue_mask = im2_data[:,:,2] == rgb[2]
    
    pixel_mask = reds_mask & green_mask & blue_mask
    
    npixels = len(im2_data[pixel_mask])
    number_of_pixels_of_a_color_2.append(npixels)
    color_labels_2.append("Color #" + str(icolor))
    color_values_2.append(rgb/255)
```

```python
for icolor, rgb in enumerate(unique_3_colors):
    reds_mask = im3_data[:,:,0] == rgb[0]
    green_mask = im3_data[:,:,1] == rgb[1]
    blue_mask = im3_data[:,:,2] == rgb[2]
    
    pixel_mask = reds_mask & green_mask & blue_mask
    
    npixels = len(im3_data[pixel_mask])
    number_of_pixels_of_a_color_3.append(npixels)
    color_labels_3.append("Color #" + str(icolor))
    color_values_3.append(rgb/255)
```

drop out the first color which is the black background so we could compare other main colors directly.

Show three graphs.

```python
fig, ax = plt.subplots(figsize=(20,5))

myBarChart = ax.bar(color_labels_1, number_of_pixels_of_a_color_1)

for i in range(len(myBarChart)):
    myBarChart[i].set_color(color_values_1[i])
    myBarChart[i].set_edgecolor('black')

plt.show()
```
